```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
test_weights = [3331,3331,1105,913,2577,526]
names = ["flat", "gable", "hip", "mixed_w_wind_credit", "mixed_wo_wind_credit", "unknown"]
plt.bar(names, test_weights)
y_pos = range(len(names))
plt.bar(y_pos, test_weights)
# Rotation of the bars names
plt.xticks(y_pos, names, rotation=90)
plt.title("test set class distribution")
```

```python
train_weights = [5269, 10732, 5946, 5730, 13469, 2822]
plt.bar(names, test_weights)
y_pos = range(len(names))
plt.bar(y_pos, train_weights)
# Rotation of the bars names
plt.xticks(y_pos, names, rotation=90)
plt.title("train set class distribution")
```
